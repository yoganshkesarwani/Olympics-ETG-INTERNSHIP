```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = pd.read_csv('summer.csv')
df
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
len(df["City"].unique())

```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python

df[df.Medal == 'Gold'].groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5))

```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]

for Sport in df["Sport"].unique():
  data.append([Sport , len(df[df["Sport"]==Sport])])

pd.DataFrame(data,columns = ['Sport','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5))

```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]

for Athlete in df["Athlete"].unique():
  data.append([Athlete , len(df[df["Athlete"]==Athlete])])

pd.DataFrame(data,columns = ['Athlete','freq']).sort_values(by='freq', ascending=False).head().plot(x = 'Athlete', y = 'freq', kind = 'bar', figsize = (10,5))

```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
df[df.Medal == 'Gold'].groupby('Athlete').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df[(df.Medal == 'Gold') & (df.Country=='IND')].Year.head(1)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby('Event').count()['Athlete'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df[(df.Medal == 'Gold') & (df.Gender=='Women')].groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot(x = 'Sport', y = 'freq', kind = 'bar', figsize = (10,5))
```
